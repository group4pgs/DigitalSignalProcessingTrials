```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
x = np.array([0.1,0.2,0.1,0.24,0.12,0.02,0.2,0.15,0.38,0.2])
b, a = np.array([1,-1.0281,0.3651]), np.array([0.0842,0.1684,0.0842])
y = signal.lfilter(b,a,x)
print(x)
print(y)
```

```python
b,a = signal.butter(2,0.3,'lowpass')
```

```python
w,h = signal.freqs(b,a)
```

```python
plt.plot(w,np.abs(h))
```
